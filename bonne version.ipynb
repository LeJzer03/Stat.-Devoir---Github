## Estimation de densité

L'histogramme que vous avez représenté peut vous aider à postuler un modèle pour la distribution de la variable étudiée. Cependant, il est nécessaire de fixer le nombre de bins, et le graphique obtenu peut être difficile à interpréter, la visualisation dépendant du choix du nombre de bins.

La construction d'un tel histogramme peut être vue comme un empilement de blocs, chaque bloc correspondant à un point (*i.e.* une observation). Pour chaque observation se trouvant dans la plage de valeurs d'une bin, un bloc est ajouté pour cette bin. Une autre manière de construire une estimation de la densité consiste à ajouter un bloc centré en la valeur de l'observation à laquelle il correspond. Pour chaque observation, on place un bloc, non pas dans une bin arbitraire, mais à la position de l'observation sur l'axe des observations. Ensuite, on somme la hauteur des blocs en chaque point de cet axe. 

Derrière cette intuition se cache la méthode de l'estimation de la densité basée sur un noyau (*Kernel Density Estimation*, KDE, en anglais). Le noyau représente le bloc que l'on place pour chaque observation. Il peut prendre beaucoup de formes, et ne se limite donc pas à un "bloc" rectangulaire.

Cette méthode est par exemple implémentée dans la librairie [scikit-learn](https://scikit-learn.org/1.5/modules/density.html).

<div class="alert-success">

À l'aide de cette librairie, générez une figure superposant l'histogramme généré précédemment, et les densités estimées avec un noyau gaussien, pour les valeurs du paramètre "bandwidth" ($bw$) suivantes: $bw \in \{0.1, 0.3, 0.8, 1.2\}$. Faites également apparaître les données d'échantillon comme proposé dans l'exemple de l'url ci-dessus. Vous pouvez laisser les autres paramètres par défaut de la méthode de la librairie.

</div>

````python

# Convertir les données en tableau NumPy
data_np = np.array(data)

# Superposition de l'histogramme et des densités estimées avec un noyau gaussien pour différentes valeurs de bandwidth
fig, ax = plt.subplots(figsize=(10, 6))

# Histogramme
sns.histplot(data_np, bins=15, kde=False, stat="density", ax=ax, label='Histogramme')

# Estimation de la densité avec différents bandwidths
bandwidths = [0.1, 0.3, 0.8, 1.2]
colors = ['blue', 'green', 'red', 'purple']
x_vals = np.linspace(data_np.min(), data_np.max(), 1000)

for bw, color in zip(bandwidths, colors):
    kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(data_np[:, None])
    log_dens = kde.score_samples(x_vals[:, None])
    ax.plot(x_vals, np.exp(log_dens), label=f'KDE bw={bw}', color=color)

# Ajout des croix avec décalage horizontal et vertical
x_jitter = data_np + 0.05 * np.random.uniform(-0.5, 0.5, size=len(data_np))  # Décalage horizontal
y_jitter = -0.07 + 0.07 * np.random.uniform(-0.5, 0.5, size=len(data_np))  # Décalage vertical

# Sélectionner 1 point sur 10
indices = np.arange(len(data_np))
selected_indices = indices[::10]

ax.scatter(x_jitter[selected_indices], y_jitter[selected_indices], marker='x', color='black', label='Points échantillonnés', alpha=0.5)

# Configuration du graphique
ax.set_title('Histogramme et KDE avec différentes largeurs de bande')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Densité')
ax.legend()

plt.show()

````

> VOTRE TEXTE ICI

L'histogramme et l'estimation de densité par noyau (KDE) sont deux méthodes pour visualiser la distribution d'une variable. Nous allons comparer ces deux méthodes.

- Histogramme :

La méthode consiste a diviser les données en un certain nombre de bins (ici 15) et ensuite de compter le nombre de points dans chaque bin.
La hauteur de chaque bin représente la densité des données dans cette plage.
La forme de l'histogramme peut être influencée par le choix du nombre de bins, ce qui peut rendre l'interprétation difficile si ce nombre n'est pas bien choisi.

- KDE :

La méthode utilise un noyau (ici gaussien) pour estimer la densité de probabilité des données.
Chaque point de données est remplacé par une fonction de densité (le noyau) centrée sur ce point.
La somme de ces fonctions de densité donne une estimation lisse de la densité de probabilité.
La KDE est moins sensible aux choix arbitraires comme le nombre de bins, mais dépend du paramètre "bandwidth".

Le paramètre "bandwidth" (largeur de bande) contrôle la largeur du noyau utilisé pour estimer la densité. En d'autres termes, il détermine la "lissage" de la courbe KDE.

Dans le code ci-dessus, différentes valeurs de bandwidth (0.1, 0.3, 0.8, 1.2) sont utilisées pour montrer comment la KDE change avec ce paramètre.
Voici les effets observés :

- $bw=0.1$   :  La courbe KDE est très détaillée et suit de près les données, montrant beaucoup de pics et de creux.
- $bw=0.3 $  :  La courbe KDE est encore détaillée mais commence à lisser certaines des variations locales.
- $bw=0.8 $  :  La courbe KDE est plus lisse, montrant une estimation plus générale de la densité.
- $bw=1.2 $  :  La courbe KDE est très lisse, fournissant une vue très générale de la distribution, mais peut masquer des détails importants.

En résumé, le choix du bandwidth est crucial pour obtenir une estimation de densité qui est à la fois informative et représentative de la distribution sous-jacente des données. Un bandwidth trop petit peut rendre la courbe trop bruitée, tandis qu'un bandwidth trop grand peut lisser trop de détails.

<div class="alert-success">

Afin d'étudier plus en détails les propriétés du paramètre "bandwidth", générez une nouvelle figure en suivant la méthode suivante.

Pour chaque bandwidth $bw \in \{0.1, 0.3, 0.5, 1.2\}$ :

- générez 25 sous-échantillons de taille $n=25$ de l'échantillon de départ;
- estimez, à l'aide d'une KDE et d'un kernel gaussien, pour la $bw$ considérée, la densité pour chaque sous-échantillon;
- tracez la moyenne des densités estimées, et mettez en évidence la zone comprise entre cette moyenne et $\pm$ la variance de ces densités (au moyen par exemple de la fonction `fill_between` de la librairie `matplotlib`).

</div>

````python
"""VOTRE CODE ICI"""

# Convertir les données en tableau NumPy
data_np = np.array(data)

# Générer une nouvelle figure
fig, ax = plt.subplots(figsize=(10, 6))

# Tracer l'histogramme en arrière-plan
ax.hist(data_np, bins=15, density=True, alpha=0.6, color='lightblue', edgecolor='black', label='Histogram')

# Estimation de la densité avec différents bandwidths
bandwidths = [0.1, 0.3, 0.5, 1.2]
colors = ['blue', 'green', 'red', 'purple']
x_d = np.linspace(data_np.min(), data_np.max(), 1000)[:, None]

for bw, color in zip(bandwidths, colors):
    densities = []
    
    # Générer 25 sous-échantillons et estimer la densité
    for _ in range(25):
        subsample = np.random.choice(data_np, size=25, replace=False)
        kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(subsample[:, None])
        log_dens = kde.score_samples(x_d)
        densities.append(np.exp(log_dens))
    
    densities = np.array(densities)
    mean_density = densities.mean(axis=0)
    std_density = densities.std(axis=0)
    
    # Tracer la moyenne des densités estimées
    ax.plot(x_d[:, 0], mean_density, label=f'Moyenne KDE bw={bw}', color=color)
    
    # Mettre en évidence la zone comprise entre la moyenne et ± la variance
    ax.fill_between(x_d[:, 0], mean_density - std_density, mean_density + std_density, color=color, alpha=0.3)

# Ajout des croix avec décalage horizontal et vertical
x_jitter = data_np + 0.05 * np.random.uniform(-0.5, 0.5, size=len(data_np))  # Décalage horizontal
y_jitter = -0.07 + 0.07 * np.random.uniform(-0.5, 0.5, size=len(data_np))  # Décalage vertical

# Sélectionner 1 point sur 10
indices = np.arange(len(data_np))
selected_indices = indices[::10]

ax.scatter(x_jitter[selected_indices], y_jitter[selected_indices], marker='x', color='black', label='Points échantillonnés', alpha=0.5)

# Configuration du graphique
ax.set_title('Histogramme normalisé avec KDE')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Densité')
ax.legend()

plt.tight_layout()
plt.show()
````

Les résultats montrent comment la largeur de bande influence l'estimation de densité. 


Les critères modifiés par le choix du bandwidth sont les suivants :

- Détail vs. Lissage : des petites bandwidth montre plus de détails tandis que des grandes bandwidth lisse davantage.
- Sensibilité aux sous-échantillons : les petites bandwidth sont plus sensible aux variations aléatoires.
- Interprétation : les petites bandwidth peuvent être bruitées mais les grandes bandwidth peuvent masquer des détails.

Dans notre cas, prendre un bandwidth = 1.2 serait complètement inutile car cela représente un lissage beaucoup trop important qui fausse nos données.
De même, prendre un bandwidth = 0.1 détaillerait trop nos résultats.

1) Lorsque la bandwidth est plus grande ,elle retient un plus grand nombre de données dans son intervalle. Une donnée individuelle a donc moins d'effet sur la variance de l'estimation de cette bande.
2) Cependant si la bandwidth diminue, cet impacte sur la variance de l'estimation qu'une donnée individuelle possède croit drastiquement. Ce qui se traduit par l'augmentation du "nuage" autour des estimations. 

Le choix du bandwidth est un compromis entre montrer suffisamment de détails sans introduire trop de bruit, assurer la stabilité de l'estimation et faciliter l'interprétation. Il dépend du contexte et des objectifs de l'analyse.

# Modèle et estimation des paramètres


Dans cette deuxième partie du devoir, vous allez devoir construire différents estimateurs ponctuels pour le paramètre d'un modèle statistique de vos données. Pour cela, vous utiliserez comme modèle statistique la distribution Exponentielle  de paramètre $\lambda$, décalée de 6.5 unités vers la droite, dont la PDF est donnée par
$$f_Y(y ; \lambda) = \begin{cases} \lambda e^{-\lambda (y - 6.5)} , & \text{si $y \geq 6.5$} ,\\ 0, & \text{si $y < 6.5$}, \end{cases}$$
avec $\lambda>0$.

Intéressons nous à plusieurs méthodes pour estimer la valeur du paramètre de la distribution.

**Remarque importante**: 

Lorsque vous calculez des statistiques en utilisant une fonction implémentée dans une librairie Python (ou autre), vérifiez dans la documentation que cette fonction calcule bien la statistique souhaitée, pour la distribution considérée, telles que définies dans le cours théorique ou dans l'énoncé.

## Estimation par la méthode des moments

Une première manière pour estimer les paramètres de la distribution Exponentielle est d'utiliser la méthode des moments. Pour rappel, cette méthode consiste à faire coincider le moment théorique et le moment d'échantillon d'ordre $k$ :
$$E(Y^k) = \dfrac{1}{n} \sum_{j=1}^n Y_j^k .$$

### Calcul de l'estimateur

<div class="alert-success">

Calculez la valeur de l'estimateur $\hat{\lambda}_{\mathrm{MoM}}$ pour votre échantillon.

</div>

````python
"""VOTRE CODE ICI"""
# Calcul de l'estimateur par la méthode des moments
# La méthode des moments consiste à faire coïncider le moment théorique et le moment d'échantillon d'ordre k.
# Pour une distribution exponentielle décalée de 6.5 unités vers la droite, le moment d'ordre 1 (la moyenne) est donné par :
# E(Y) = 1/λ + 6.5

# Calcul de la moyenne des données observées
mean_data = data.mean()

# Utilisation de la méthode des moments pour estimer λ
# En égalisant la moyenne observée à la moyenne théorique, nous avons :
# mean_data = 1/λ + 6.5
# En résolvant pour λ, nous obtenons :
# λ = 1 / (mean_data - 6.5)
lambda_mom = 1 / (mean_data - 6.5)

# Affichage de l'estimateur de λ
print(f"Estimateur MoM de lambda: {lambda_mom}")
````

## Estimation par la méthode du maximum de vraisemblance


Une seconde méthode d'estimation des paramètres est la méthode du maximum de vraisemblance. Pour rappel, cette méthode consiste à trouver la valeur des paramètres maximisant la fonction de vraisemblance (ou, de manière équivalente, maximisant la log-vraisemblance) :
$$ \hat{\lambda}_\mathrm{MLE} = \underset{\lambda}{\mathrm{arg\,max}} \; \log L(\lambda; \textbf{y}),$$
où $\textbf{y}$ représente les données observées.

*Note :* la notation $\log$ dénote dans la convention du cours le logarithme naturel.

### Calcul des estimateurs

<div class="alert-success">

Calculez la valeur de l'estimateur $\hat{\lambda}_{\mathrm{MLE}}$ pour votre échantillon.

</div>

````python
"""VOTRE CODE ICI"""
def log_likelihood(lambda_val):
    # Calcul de la log-vraisemblance
    return len(data) * np.log(lambda_val) - lambda_val * np.sum(data - 6.5)

# Maximisation de la log-vraisemblance
# La fonction minimize_scalar permet de minimiser une fonction unidimensionnelle
# En multipliant la log-vraisemblance par -1, nous obtenons le maximum de la log-vraisemblance
#Nous allons chercher le maximum de la log-vraisemblance dans l'intervalle [1e-6, 10], arbitrairement choisi
result = minimize_scalar(lambda lambda_val: -log_likelihood(lambda_val), bounds=(1e-6, 10), method='bounded')

# Résultat : estimateur MLE
lambda_mle = result.x
print(f"Estimateur MLE de lambda: {lambda_mle}")
````

## Biais et variance de l'estimateur

Afin d'étudier le concept de biais et d'estimer le biais de l'estimateur MLE, vous allez premièrement comparer les effets de certains paramètres sur la précision de cet estimateur.

<div class="alert-success">

Pour ce faire, vous allez générer les courbes suivantes, en tirant $m=500$ échantillons différents :

- dans une première figure, l'évolution du **biais** (en valeur absolue) de $\hat{\lambda}_{\mathrm{MLE}}$ en fonction de la taille d'échantillon $n$, pour $n \in \{2, 3, ..., 100\}$
- dans une seconde figure, l'évolution de la **variance** de $\hat{\lambda}_{\mathrm{MLE}}$ en fonction de la taille d'échantillon $n$, pour $n \in \{2, 3, ..., 100\}$.

Au total, deux figures sont attendues, chacune contenant une courbe. Pensez à ajouter une légende ou un titre sur vos deux figures, afin d'éviter une confusion entre les courbes (en plus des labels d'axes, évidemment).

````python
"""VOTRE CODE ICI"""

# Initialisation des listes pour stocker le biais et la variance
biases = []
variances = []

# Valeur théorique de lambda
lambda_theorique = 2.272

# Taille des échantillons
n_values = range(2, 101)
n_values = np.array(n_values)


# Génération des échantillons et calcul du biais et de la variance
for n in n_values:
    estimates = []
    for _ in range(500):
        sample = np.random.exponential(1/lambda_theorique, n) + 6.5
        lambda_mle_estimate = n / np.sum(sample - 6.5)
        estimates.append(lambda_mle_estimate)
    
    estimates = np.array(estimates)
    bias = np.abs(np.mean(estimates) - lambda_theorique)
    variance = np.var(estimates)
    
    biases.append(bias)
    variances.append(variance)

plt.figure(figsize=(10, 6))
plt.plot(n_values, biases, label="Biais de $\\hat{\\lambda}_{MLE}$")
plt.xlabel('Taille de l\'échantillon (n)')
plt.ylabel('Biais')
plt.title('Évolution du biais de l\'estimateur MLE ($\\hat{\\lambda}_{MLE}$) en fonction de la taille de l\'échantillon')
plt.legend()
plt.minorticks_on()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.tick_params(which='both', direction='in', top=True, right=True)

plt.show()

# Tracé de la variance
plt.figure(figsize=(10, 6))
plt.plot(n_values, variances, label="Variance de $\\hat{\\lambda}_{MLE}$")
plt.xlabel('Taille de l\'échantillon (n)')
plt.ylabel('Variance')
plt.title('Évolution de la variance de l\'estimateur MLE ($\\hat{\\lambda}_{MLE}$) en fonction de la taille de l\'échantillon')
plt.legend()
plt.minorticks_on()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.tick_params(which='both', direction='in', top=True, right=True)

plt.show()
````

Le biais d'un estimateur est la différence entre la valeur attendue de l'estimateur et la vraie valeur du paramètre.$\\$
Lorsque la taille des échantillons augmente, le biais de l'estimateur du maximum de vraisemblance (MLE) tend à diminuer. En d'autres termes, l'estimateur devient de plus en plus précis et converge vers la vraie valeur du paramètre $\lambda$.

La variance d'un estimateur, quant à elle, mesure la dispersion des valeurs de l'estimateur autour de sa moyenne.$\\$
Lorsque la taille des échantillons augmente, la variance de l'estimateur du maximum de vraisemblance (MLE) tend également à diminuer. Cela signifie que les estimations deviennent de plus en plus concentrées autour de la vraie valeur du paramètre $\lambda$. $\\$

 - Explication théorique $\\$

L'estimateur du maximum de vraisemblance est consistant, ce qui signifie qu'il converge en probabilité vers la vraie valeur du paramètre à mesure que la taille de l'échantillon augmente. En augmentant la taille de l'échantillon, nous obtenons plus d'informations sur la distribution sous-jacente, ce qui permet à l'estimateur MLE de devenir de plus en plus précis. $\\$

L'estimateur MLE est asymptotiquement non biaisé, ce qui signifie que le biais de l'estimateur tend vers zéro à mesure que la taille de l'échantillon augmente.
A l'aide d'un grand nombre d'observations, l'estimateur MLE utilise suffisamment d'informations pour fournir une estimation précise du paramètre $\lambda$. $\\$

La variance de l'estimateur MLE diminue à mesure que la taille de l'échantillon augmente, et elle est asymptotiquement égale à l'inverse de l'information de Fisher.

Pour obtenir un estimateur non-biaisé de $\lambda$ à partir de l'estimateur du maximum de vraisemblance $\hat{\lambda}_{\mathrm{MLE}}$, il suffit de multiplier cet estimateur $\hat{\lambda}_{\mathrm{MLE}}$ par $\frac{n-1}{n}$. $\\$

Ce qui s'explique mathématiquement par le résonnement suivant : $\\$

L'estimateur du maximum de vraisemblance pour une distribution exponentielle décalée est donné par : $$\hat{\lambda}_{\mathrm{MLE}} = \dfrac{n}{\sum_{i=1}^n (y_i - 6.5)}$$

Cet estimateur est biaisé pour des échantillons de taille finie. Pour obtenir un estimateur non-biaisé, nous utilisons la correction suivante :

$$\hat{\lambda}_{\mathrm{unbiased}} = \dfrac{n-1}{\sum_{i=1}^n (y_i - 6.5)}$$

Ce qui rend bien la relation,

$$ \hat{\lambda}_{\mathrm{unbiased}} =  \frac{n-1}{n} \hat{\lambda}_{\mathrm{MLE}}  $$


La correction du facteur $n$ par $n-1$ est utilisée pour compenser le biais introduit par l'estimation de la moyenne à partir de l'échantillon. Cette correction est similaire à celle utilisée pour obtenir une estimation non-biaisée de la variance dans un échantillon.

````python
"""VOTRE CODE ICI"""

import numpy as np
import matplotlib.pyplot as plt

# Définir la valeur théorique de lambda
lambda_theorique = 2.272

# Différentes tailles d'échantillon à tester
sample_sizes = [5, 25, 100, 200]

# Nombre de répétitions pour chaque taille d'échantillon
num_repetitions = 500

# Fonction pour générer les échantillons et calculer les estimateurs
def generate_estimates(sample_size, num_repetitions, lambda_theorique):
    mle_estimates = []
    unbiased_estimates = []
    mean_estimates = []
    
    for _ in range(num_repetitions):
        sample = np.random.exponential(1/lambda_theorique, sample_size) + 6.5
        mle_estimate = sample_size / np.sum(sample - 6.5)
        unbiased_estimate = (sample_size - 1) / np.sum(sample - 6.5)
        mean_estimate = 1 / mle_estimate 
        
        mle_estimates.append(mle_estimate)
        unbiased_estimates.append(unbiased_estimate)
        mean_estimates.append(mean_estimate)
    
    return np.array(mle_estimates), np.array(unbiased_estimates), np.array(mean_estimates)

# Générer les histogrammes pour chaque taille d'échantillon
colors = ['red', 'green', 'blue', 'purple']
labels = [f'n={n}' for n in sample_sizes]

# Historique des paramètres d’affichage des graphiques
def plot_histogram(estimates, est_type, theoretical_value, xlabel):
    plt.figure(figsize=(12, 8))
    for i, sample_size in enumerate(sample_sizes):
        plt.hist(estimates[sample_size], bins=20, alpha=0.5, label=labels[i], color=colors[i], density=True)

    # Superposition de deux lignes pour le style multicolore de la ligne théorique
    plt.axvline(theoretical_value, color='yellow', linestyle='-', linewidth=2,)
    plt.axvline(theoretical_value, color='red', linestyle='--', linewidth=2, label='Valeur théorique')

    plt.xlabel(xlabel)
    plt.ylabel('Densité')
    plt.title(f'Histogramme de l’estimateur {est_type} pour Différentes Tailles d’Échantillon')
    plt.legend()
    plt.show()

# Générer et afficher les graphiques pour chaque type d’estimateur
estimates = {n: generate_estimates(n, num_repetitions, lambda_theorique) for n in sample_sizes}

# Histogramme pour l'estimateur MLE
plot_histogram({n: estimates[n][0] for n in sample_sizes}, 'MLE', lambda_theorique, 'Estimateur de λ (MLE)')

# Histogramme pour l'estimateur non-biaisé
plot_histogram({n: estimates[n][1] for n in sample_sizes}, 'Non-Biaisé', lambda_theorique, 'Estimateur de λ (Non-Biaisé)')

# Histogramme pour l'estimateur de la moyenne
theoretical_mean = 1 / lambda_theorique 
plot_histogram({n: estimates[n][2] for n in sample_sizes}, 'Moyenne', theoretical_mean, 'Estimateur de la Moyenne (1/λ)')
````
