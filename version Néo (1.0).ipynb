## Estimation de densité

L'histogramme que vous avez représenté peut vous aider à postuler un modèle pour la distribution de la variable étudiée. Cependant, il est nécessaire de fixer le nombre de bins, et le graphique obtenu peut être difficile à interpréter, la visualisation dépendant du choix du nombre de bins.

La construction d'un tel histogramme peut être vue comme un empilement de blocs, chaque bloc correspondant à un point (*i.e.* une observation). Pour chaque observation se trouvant dans la plage de valeurs d'une bin, un bloc est ajouté pour cette bin. Une autre manière de construire une estimation de la densité consiste à ajouter un bloc centré en la valeur de l'observation à laquelle il correspond. Pour chaque observation, on place un bloc, non pas dans une bin arbitraire, mais à la position de l'observation sur l'axe des observations. Ensuite, on somme la hauteur des blocs en chaque point de cet axe. 

Derrière cette intuition se cache la méthode de l'estimation de la densité basée sur un noyau (*Kernel Density Estimation*, KDE, en anglais). Le noyau représente le bloc que l'on place pour chaque observation. Il peut prendre beaucoup de formes, et ne se limite donc pas à un "bloc" rectangulaire.

Cette méthode est par exemple implémentée dans la librairie [scikit-learn](https://scikit-learn.org/1.5/modules/density.html).

<div class="alert-success">

À l'aide de cette librairie, générez une figure superposant l'histogramme généré précédemment, et les densités estimées avec un noyau gaussien, pour les valeurs du paramètre "bandwidth" ($bw$) suivantes: $bw \in \{0.1, 0.3, 0.8, 1.2\}$. Faites également apparaître les données d'échantillon comme proposé dans l'exemple de l'url ci-dessus. Vous pouvez laisser les autres paramètres par défaut de la méthode de la librairie.

</div>

````python
from scipy.integrate import simps
bandwidths = [0.1, 0.3, 0.8, 1.2]

magnitude = np.linspace(data.min() - 3, data.max() + 3, 1000).reshape(-1, 1)
for bw in bandwidths:
    kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(data.values.reshape(-1, 1))
    log_dens = kde.score_samples(magnitude)
    plt.plot(magnitude, np.exp(log_dens), label=f"KDE (bandwidth={bw})")
    integral = simps(np.exp(log_dens), magnitude.flatten())  
    print(f"Intégrale de la KDE avec bandwidth={bw} : {integral:.4f}")
    

plt.scatter(data, np.full_like(data, -0.1), marker='+', color='black', label="Échantillon de données", linewidths=1)

plt.xlabel("magnitude")
plt.ylabel("Densité")
plt.title("Histogramme et estimation de densité par noyau")
plt.legend()
plt.show()
````

Afin d'étudier plus en détails les propriétés du paramètre "bandwidth", générez une nouvelle figure en suivant la méthode suivante.

Pour chaque bandwidth $bw \in \{0.1, 0.3, 0.5, 1.2\}$ :

- générez 25 sous-échantillons de taille $n=25$ de l'échantillon de départ;
- estimez, à l'aide d'une KDE et d'un kernel gaussien, pour la $bw$ considérée, la densité pour chaque sous-échantillon;
- tracez la moyenne des densités estimées, et mettez en évidence la zone comprise entre cette moyenne et $\pm$ la variance de ces densités (au moyen par exemple de la fonction `fill_between` de la librairie `matplotlib`).

````python
np.random.seed(0)  
bandwidths = [0.1, 0.3, 0.5, 1.2]
n_samples = 25
plt.figure(figsize=(15, 8))

for bw in bandwidths:
    density_estimates = []
    
    for _ in range(25):
        sample = np.random.choice(data, size=n_samples, replace=True)
        kde = KernelDensity(kernel='gaussian', bandwidth=bw).fit(sample.reshape(-1, 1))
        log_dens = kde.score_samples(magnitude)
        density_estimates.append(np.exp(log_dens))
    
    density_estimates = np.array(density_estimates)
    mean_density = density_estimates.mean(axis=0)
    variance_density = density_estimates.var(axis=0)

    plt.plot(magnitude, mean_density, label=f"Moyenne KDE (bandwidth={bw})")
    
    plt.fill_between(magnitude.flatten(),
                     mean_density - np.sqrt(variance_density),
                     mean_density + np.sqrt(variance_density),
                     alpha=0.2)

plt.scatter(data, np.full_like(data, -0.05), marker='+', color='black', label="Échantillon de données", linewidths=1)
plt.xlabel("Magnitude")
plt.ylabel("Densité")
plt.title("Estimation de densité par noyau avec sous-échantillons")
plt.legend()
plt.show()
````

En effet lorsque la largeur de la bande passante augmente, cette dernière retient un plus grand nombre de données dans son intervalle et par conséquent l'effet d'une donnée individuelle ne peut avoir que moins d'impacte sur la variance de l'estimation de cette bande. A l'inverse lorsque la bandwidth diminue, chaque donnée a un plus grand impacte sur la variance d'où l'augmentation du "nuage" autour des estimations. 

> Le choix de cette valeur résulte d'un compromis. En effet, une bandwidth plus petite peut conduire à un surajustement, ce qui pourrait capturer le bruit présent dans les données et ainsi rendre la une estimation de densité irrégulière. En revenche, une bandwidth trop large conduirais à des sous-ajustements, résultant en un lissage des données. La bandwidth doit par conséquent être adaptée aux différentes situations rencontrées en tenant compte de ce compromis (biais-variance).

# Modèle et estimation des paramètres


Dans cette deuxième partie du devoir, vous allez devoir construire différents estimateurs ponctuels pour le paramètre d'un modèle statistique de vos données. Pour cela, vous utiliserez comme modèle statistique la distribution Exponentielle  de paramètre $\lambda$, décalée de 6.5 unités vers la droite, dont la PDF est donnée par
$$f_Y(y ; \lambda) = \begin{cases} \lambda e^{-\lambda (y - 6.5)} , & \text{si $y \geq 0$} ,\\ 0, & \text{si $y < 0$}, \end{cases}$$
avec $\lambda>0$.

Intéressons nous à plusieurs méthodes pour estimer la valeur du paramètre de la distribution.

Lorsque vous calculez des statistiques en utilisant une fonction implémentée dans une librairie Python (ou autre), vérifiez dans la documentation que cette fonction calcule bien la statistique souhaitée, pour la distribution considérée, telles que définies dans le cours théorique ou dans l'énoncé.

## Estimation par la méthode des moments

Une première manière pour estimer les paramètres de la distribution Exponentielle est d'utiliser la méthode des moments. Pour rappel, cette méthode consiste à faire coincider le moment théorique et le moment d'échantillon d'ordre $k$ :
$$E(Y^k) = \dfrac{1}{n} \sum_{j=1}^n Y_j^k .$$

### Calcul de l'estimateur

<div class="alert-success">

Calculez la valeur de l'estimateur $\hat{\lambda}_{\mathrm{MoM}}$ pour votre échantillon.

</div>

````python
data_mean = np.mean(data)

# ordre 1: E(Y) = Y_mean
# i.e. E(distribution exponentielle) = Y_mean, posons Y = X + 6.5 
# puisque la distribution Y est une distribution lanbda.exp(-labmda.y) décalée de 6.5 unités)
# où E(X + 6.5) = E(X) + 6.5 = 1/lambda + 6.5

lambda_mom = 1/(mean_val - 6.5)
print("Estimateur de lambda par la méthode des moments :", lambda_mom)

#Question : doit on se tracasser du fait que la fonction fY doit etre = 0 pour y < 0 ?
````

## Estimation par la méthode du maximum de vraisemblance


Une seconde méthode d'estimation des paramètres est la méthode du maximum de vraisemblance. Pour rappel, cette méthode consiste à trouver la valeur des paramètres maximisant la fonction de vraisemblance (ou, de manière équivalente, maximisant la log-vraisemblance) :
$$ \hat{\lambda}_\mathrm{MLE} = \underset{\lambda}{\mathrm{arg\,max}} \; \log L(\lambda; \textbf{y}),$$
où $\textbf{y}$ représente les données observées.

*Note :* la notation $\log$ dénote dans la convention du cours le logarithme naturel.

### Calcul des estimateurs

<div class="alert-success">

Calculez la valeur de l'estimateur $\hat{\lambda}_{\mathrm{MLE}}$ pour votre échantillon.

</div>

````python
lambda_mle = len(data) / np.sum(data - 6.5)

print(f"Valeur de l'estimateur du maximum de vraisemblance pour λ : {lambda_mle}")
````

## Comparaison

<div class="alert-success">

Maintenant que les valeurs des estimateurs ont été déterminées, superposez les modèles estimés $f_Y(y ; \hat{\lambda})$ par la méthode des moments et du maximum de vraisemblance à l'histogramme de vos données d'échantillon $\mathbf{y} = (y_1, \ldots, y_n)$.

</div>

````python

def pdf(y,lambda_hat):
    if y<6.5 :
        return 0
    else : 
        return lambda_hat*np.exp(-lambda_hat*(y-6.5)) 
    
y = np.linspace(6.5,data.max()+1,1000)
pdf_mom = np.zeros_like(y)
pdf_mle = np.zeros_like(y)
for i in range(len(y)): 
    pdf_mom[i] = pdf(y[i], lambda_mom)
    pdf_mle[i] = pdf(y[i], lambda_mle)
    
plt.plot(y, pdf_mom, label=f"MoM : $\\lambda={lambda_mom:.2f}$", color='red', linestyle='--')
plt.plot(y, pdf_mle, label=f"MLE : $\\lambda={lambda_mle:.2f}$", color='green', linestyle='-.')
plt.hist(data, bins=15, density=True, alpha=0.8, color='skyblue', edgecolor='black')


plt.xlabel("Valeurs de Y")
plt.ylabel("Densité")
plt.title("Modèles estimés (MoM et MLE)")
plt.legend()
plt.show()
````

````python
lambda_th = 2.272
m = 500
n = np.arange(2, 101)

biais = np.zeros_like(n, float)
variance = np.zeros_like(n, float)
lambda_est = np.zeros(m, float)

for i in range(len(n)):
    for j in range(m):
        sample = np.random.exponential(1/lambda_th, n[i]) + 6.5 
        lambda_est[j] = n[i]/np.sum(sample-6.5)
    
    biais[i] = np.abs(np.mean(lambda_est) - lambda_th)
    variance[i] = np.mean((lambda_est - np.mean(lambda_est))**2)

plt.plot(n, biais, label="Biais de $\\hat{\\lambda}_{MLE}$", color='red')
plt.xlabel("Taille de l'échantillon (n)")
plt.ylabel("Biais (valeur absolue)")
plt.title("Evolution du biais de l'estimateur MLE en fonction de la taille d'échantillon")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(n, variance, label="Variance de $\\hat{\\lambda}_{MLE}$", color='blue')
plt.xlabel("Taille de l'échantillon (n)")
plt.ylabel("Variance")
plt.title("Evolution de la variance de l'estimateur MLE en fonction de la taille d'échantillon")
plt.legend()
plt.grid(True)
plt.show()
````

de l'estimateur, lorsque la taille $n$ des échantillons augmente. Pouvez-vous en donner une explication théorique ? En particulier, quelle propriété de l'estimateur du maximum de vraisemblance est ici observée ?

</div>

> La valeur du biais et de la variance de l'estimateur diminiue jusqu'à tendre vers 0 avec l'augmentation de n. Ces observations peuvent être expliquées par la consistance d'un estimateur. En effet, un estimateur $\hat{\theta}$ est consistant si la probabilité que la différence entre cet estimateur et le vrai paramètre soit arbitrairement grande tende vers 0 pour un taille d'échantillon n tendant vers l'infini. La condition suffisante pour que $\hat{\theta}$ soit un estimateur consistant de $\theta$ étant que $MSE(\hat{\theta})$ tende vers 0 si n tend vers l'infini. On en déduit, par la définition de $MSE(\hat{\theta})$, que $bias(\hat{\theta})$ et $Var(\hat{\theta})$ tendent vers 0 si n tend vers l'infini.
