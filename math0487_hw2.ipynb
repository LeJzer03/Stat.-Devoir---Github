## Approximation Normale

Vous allez ici construire un intervalle de confiance *approximé*. En effet, il vous est demandé de faire une approximation Normale, *i.e.* de considérer que la distribution de l'estimateur considéré est approximativement Normale.

Sans hypothèse quant à la distribution de $Vmag$ dans la population, une telle approximation ne peut cependant être faite qu'à certaine(s) condition(s). On suppose les données d'échantillons indépendantes et identiquement distribuées.

<div class="alert alert-block alert-success">

Quelle(s) condition(s) doive(nt) être remplie(s) ? Quel théorème cette(ces) condition(s) permet(tent)-elle(s) d'appliquer ?

</div>



>Réponse :
Pour appliquer une approximation normale à l'estimateur (ici la moyenne de 
Vmag), deux conditions doivent être remplies :

Les données doivent être indépendantes et identiquement distribuées (iid).
La taille de l'échantillon doit être suffisamment grande (
n élevé).
Ces conditions permettent d’appliquer le théorème central limite (TCL), qui stipule que, pour un échantillon iid suffisamment grand, la distribution de la moyenne échantillonnale suit une distribution normale, indépendamment de la distribution de la population d'origine.

Ces conditions garantissent ainsi que l'approximation normale est valable pour l'intervalle de confiance.

<div class="alert alert-block alert-success">

Calculez à présent l'intervalle de confiance pour la magnitude apparente moyenne au niveau de confiance demandé pour votre échantillon.

</div>

````python
# Calculate the mean and standard error of the mean
mean_vmag = data['Vmag'].mean()
sem_vmag = stats.sem(data['Vmag'])

# Calculate the confidence interval
confidence_level = 0.95
confidence_interval = stats.t.interval(confidence_level, sample_size - 1, loc=mean_vmag, scale=sem_vmag)

print(f"Intervalle de confiance à {confidence_level*100}% pour la magnitude apparente moyenne: {confidence_interval}")
````

<div class="alert alert-block alert-success">

Calculez maintenant des intervalles de confiance pour la magnitude apparente moyenne aux niveaux $1-\alpha \in [0.50, 0.51, 0.52, \ldots 0.97, 0.98, 0.99]$. Représentez l'évolution des bornes de ces intervalles en fonction de $1-\alpha$ sur un graphe.

</div>

````python
confidence_levels = np.arange(0.50, 1.00, 0.01)
lower_bounds = []
upper_bounds = []

mean_vmag = data['Vmag'].mean()
sem_vmag = stats.sem(data['Vmag'])

for confidence_level in confidence_levels:
    confidence_interval = stats.t.interval(confidence_level, sample_size - 1, loc=mean_vmag, scale=sem_vmag)
    lower_bounds.append(confidence_interval[0])
    upper_bounds.append(confidence_interval[1])

plt.plot(confidence_levels, lower_bounds, label='Lower Bound')
plt.plot(confidence_levels, upper_bounds, label='Upper Bound')
plt.xlabel('Confidence Level (1 - alpha)')
plt.ylabel('Bounds of Confidence Interval')
plt.title('Evolution of Confidence Interval Bounds for Apparent Magnitude Mean')
plt.legend()

# Add minor ticks and grid
plt.minorticks_on()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.gca().set_axisbelow(True)

plt.show()
````

<div class="alert alert-block alert-success">

Qu'observez-vous lorsque l'on augmente (resp. diminue) le niveau de confiance $1-\alpha$ ? Interprétez.

</div>

> Lorsque le niveau de confiance 

1−α augmente, les bornes de l'intervalle de confiance s'écartent davantage l'une de l'autre. Cela signifie que l'intervalle devient plus large. En revanche, lorsque 1−α diminue, les bornes se rapprochent, réduisant ainsi la largeur de l'intervalle.

Interprétation :
Un intervalle de confiance plus large (avec un 1−α élevé) reflète une plus grande incertitude associée à la garantie que la vraie valeur de la magnitude apparente moyenne est incluse dans l'intervalle. À l'inverse, un intervalle plus étroit (avec un 

1−α faible) implique une moindre confiance que la valeur réelle soit incluse, car on accepte un niveau d'erreur plus élevé (
α plus grand).

En résumé :

Augmenter 
1−α (niveau de confiance élevé) réduit le risque d'erreur mais élargit l'intervalle.
Diminuer 1−α (niveau de confiance faible) réduit la largeur de l'intervalle, mais au prix d'une plus grande probabilité d'exclure la vraie valeur.
Cela illustre le compromis entre précision et confiance dans l'interprétation des intervalles de confiance.

## Bootstrapping

Vous allez à présent construire un intervalle en utilisant une autre méthode que la précédente : le *bootstrapping*.

<div class="alert alert-block alert-success">

Premièrement, de manière générique, donnez *toutes* les différentes étapes de la méthode du *bootstrapping* pour construire un intervalle de confiance au niveau de confiance $1-\alpha$ pour un paramètre $\theta$ à partir d'un échantillon de taille $n$, avec un nombre d'échantillons bootstrap $m$, et en utilisant la méthode du percentile.

</div>

> .
Pour construire un intervalle de confiance au niveau de confiance $1-\alpha$ pour un paramètre $\theta$ à partir d'un échantillon de taille $n$, avec un nombre d'échantillons bootstrap $m$, et en utilisant la méthode du percentile, suivez les étapes suivantes :

1. **Échantillonnage Bootstrap** :
    - Générer $m$ échantillons bootstrap en tirant avec remise $m$ échantillons de taille $n$ à partir de l'échantillon original.

2. **Calcul de l'Estimateur** :
    - Pour chaque échantillon bootstrap, calculer l'estimateur $\hat{\theta}^*$ correspondant au paramètre $\theta$.

3. **Distribution Bootstrap** :
    - Construire la distribution empirique des $\hat{\theta}^*$ obtenus à partir des $m$ échantillons bootstrap.

4. **Intervalle de Confiance** :
    - Déterminer les percentiles $\alpha/2$ et $1-\alpha/2$ de la distribution empirique des $\hat{\theta}^*$.
    - Les valeurs correspondantes à ces percentiles forment les bornes inférieure et supérieure de l'intervalle de confiance bootstrap pour $\theta$.

5. **Interprétation** :
    - L'intervalle de confiance obtenu représente l'intervalle dans lequel le paramètre $\theta$ se trouve avec une probabilité de $1-\alpha$.

<div class="alert alert-block alert-success">

Calculez un tel intervalle au niveau de confiance précisé plus tôt pour la magnitude apparente moyenne, avec un nombre d'échantillons bootstrap $m=1000$.

</div>


````python
# Number of bootstrap samples
m = 1000

# Bootstrap sampling
bootstrap_means = []
for _ in range(m):
    bootstrap_sample = data['Vmag'].sample(n=sample_size, replace=True)
    bootstrap_means.append(bootstrap_sample.mean())

# Calculate the confidence interval
alpha = 0.05
lower_bound = np.percentile(bootstrap_means, 100 * (alpha / 2))
upper_bound = np.percentile(bootstrap_means, 100 * (1 - alpha / 2))

print(f"Intervalle de confiance bootstrap à {100 * (1 - alpha)}% pour la magnitude apparente moyenne: ({lower_bound}, {upper_bound})")
````

<div class="alert alert-block alert-success">

Générez l'histogramme des valeurs de l'estimateur pour vos échantillons bootstrap. Superposez à cet histogramme deux droites verticales afin de représenter les bornes de l'intervalle construit ci-avant.

</div>

````python
# Generate histogram of bootstrap means
plt.hist(bootstrap_means, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(lower_bound, color='red', linestyle='dashed', linewidth=2, label=f'Lower Bound: {lower_bound:.2f}')
plt.axvline(upper_bound, color='blue', linestyle='dashed', linewidth=2, label=f'Upper Bound: {upper_bound:.2f}')
plt.xlabel('Bootstrap Means of Vmag')
plt.ylabel('Frequency')
plt.title('Histogram of Bootstrap Means with Confidence Interval Bounds')
plt.legend()

# Add minor ticks and grid
plt.minorticks_on()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.gca().set_axisbelow(True)

plt.show()
````

# Régression linéaire, qualité de l'estimation, résidus de régression et leverage score

Vous allez désormais étudier le concept de régression linéaire prédictive avec les données de votre échantillon. Lorsqu'il vous est demandé d'estimer les paramètres d'un modèle de régression, vous pouvez utiliser la librairie *scikit-learn*.


Pour rappel, un modèle de regression linéaire peut être écrit sous la forme $$E(Z \mid \mathbf{X}=(x_1, \ldots, x_k), (\theta_0, \theta_1, \ldots \theta_k)) = \mu(\mathbf{x} \mid (\theta_0, \theta_1, \ldots \theta_k) ) = \theta_0 + \theta_1 x_1 + \ldots + \theta_k x_k,$$
avec $Z$ la variable aléatoire d'observation, $\mathbf{X}$ le vecteur $(X_1, \ldots, X_k)$ de variables aléatoires prédictives et $(\theta_0, \theta_1, \ldots \theta_k)$ les paramètres ou coefficients de régression (avec $\theta_0$ l'intercept).

Un ami astronome vous suggère de commencer par prédire la différence entre la magnitude absolue et la magnitude apparente (cette différence devient ainsi votre variable d'observation).

<div class="alert alert-block alert-success">

</div>

````python
# Define the variable of observation and the predictive variable
data['Mag_diff'] = data['Amag'] - data['Vmag']
X = data[['Plx']].values.reshape(-1, 1)
y = data['Mag_diff'].values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Display the coefficients of the regression
intercept = model.intercept_
coefficient = model.coef_[0]

print(f"Intercept: {intercept}")
print(f"Coefficient: {coefficient}")
````

Il vous est désormais demandé d’étudier le leverage score des différentes cristallisations.

Une manière d’obtenir le leverage score de la $j^{\text{ème}}$ observation $\mathbf{x}_j = (x_{j1}, x_{j2}, ..., x_{jK})$ est de revenir aux équations Normales de la régression.

Dans le cadre de la régression linéaire, il existe une transformation linéaire reliant les cristallisations des variables d’observation aux valeurs prédites par le modèle. Plus précisément, on note :

- $\mathbf{X}$ : la matrice contenant les cristallisations des variables prédictives 
$$ 
\mathbf{X} = 
\begin{pmatrix} 
    1 & x_{11} & \cdots & x_{1K} \\
    1 & x_{21} & \cdots & x_{2K} \\
    \vdots & \vdots  & & \vdots  \\
    1 & x_{n1} & \cdots & x_{nK} \\
\end{pmatrix} ,
$$
- $\mathbf{y} = (y_1, \ldots, y_n)^T$ : le vecteur des cristallisations des variables d’observation,
- $n$ : le nombre total de cristallisations.

Il est possible de construire un modèle de régression linéaire ajusté sur ces données. On peut alors montrer que le vecteur des prédictions $\mathbf{\hat{y}}$ s’exprime de la manière suivante : 
$$ \mathbf{\hat{y}} = \mathbf{X} (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y} = \mathbf{H} \mathbf{y} $$
où $\mathbf{H}$ est appelée la Hat matrix car elle "ajoute le chapeau" à $\mathbf{y}$.

Il est possible de trouver le leverage score de la $j^{\text{ème}}$ observation $\mathbf{x}_j$ en prenant le $j$ème élément dans la diagonale de $\mathbf{H}$.

Nous allons maintenant essayer de comprendre de manière plus intuitive ce que représente ce leverage score.

````python
# Define the initial and final positions of the point
x_new = 5
y_start = -50
y_end = 50
num_points = 100
y_new_values = np.linspace(y_start, y_end, num_points)

# Lists to store leverage scores and residuals
leverage_scores = []
residuals = []

for y_new in y_new_values:
    # Add the new point to the dataset
    data_with_new_point = data.copy()
    data_with_new_point = data_with_new_point.append({'Plx': x_new, 'Mag_diff': y_new}, ignore_index=True)
    
    # Define the predictive variable and the variable of observation
    X = data_with_new_point[['Plx']].values.reshape(-1, 1)
    y = data_with_new_point['Mag_diff'].values
    
    # Create and fit the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Calculate the leverage score for the new point
    X_with_intercept = np.hstack([np.ones((X.shape[0], 1)), X])
    hat_matrix = X_with_intercept @ np.linalg.inv(X_with_intercept.T @ X_with_intercept) @ X_with_intercept.T
    leverage_score = hat_matrix[-1, -1]
    leverage_scores.append(leverage_score)
    
    # Calculate the residual for the new point
    y_pred = model.predict(X)
    residual = y[-1] - y_pred[-1]
    residuals.append(residual)

# Plot the leverage scores and residuals
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(y_new_values, leverage_scores, label='Leverage Score')
plt.xlabel('y_new')
plt.ylabel('Leverage Score')
plt.title('Leverage Score vs y_new')
plt.legend()
plt.minorticks_on()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.gca().set_axisbelow(True)

plt.subplot(1, 2, 2)
plt.plot(y_new_values, residuals, label='Residual', color='red')
plt.xlabel('y_new')
plt.ylabel('Residual')
plt.title('Residual vs y_new')
plt.legend()
plt.minorticks_on()
plt.grid(which='both', linestyle='--', linewidth=0.5)
plt.gca().set_axisbelow(True)

plt.tight_layout()
plt.show()
````

### Interprétation de la notion de leverage :
Le leverage quantifie l'influence qu'une observation donnée exerce sur les prédictions du modèle. Mathématiquement, il correspond à la diagonale de la matrice Hat (\(H = X(X'X)^{-1}X'\)), et chaque élément de leverage représente la sensibilité des prédictions aux changements dans une observation spécifique.

- **Faible leverage** : Indique que le point est proche des autres observations dans l’espace des variables prédictives et qu’il a peu d’influence sur le modèle.
- **Fort leverage** : Correspond à un point éloigné des autres observations (souvent dans les extrémités ou en dehors de la plage des données), ce qui signifie qu'il peut fortement influencer les prédictions.

### Utilité du leverage :
Le leverage est utile pour :
- **Identifier les points influents** : Les points avec un leverage élevé peuvent avoir un impact disproportionné sur l’ajustement du modèle.
- **Diagnostiquer la qualité du modèle** : En repérant les points problématiques, on peut ajuster le modèle ou traiter les données pour améliorer la robustesse des résultats.
- **Aider dans la prise de décision** : Les points avec un leverage élevé pourraient nécessiter une attention particulière avant d'être utilisés dans des applications sensibles.

En conclusion, ces résultats montrent que le leverage est une mesure essentielle pour comprendre comment les observations influencent le modèle et pour diagnostiquer les problèmes potentiels dans l'ajustement du modèle.

# Régression logistique

Nous allons désormais utiliser un modèle de régression logistique pour résoudre un problème de classification binaire : il nous faut prédire si l'étoile est une naine ou une géante. Ceci va se faire en utilisant  deux variables prédictives : 
- l'indice de couleur *B-v*,
- la magnitude absolue *$Amag$*.

Vous pouvez utiliser la classe LogisticRegression de scikit-learn pour cette partie du devoir.

Pour rappel, un modèle de régression logistique estime la probabilité qu'une observation appartienne à une catégorie en appliquant une fonction sigmoïde sur le résultat d'une régression linéaire $$ P(Y = 1 \mid \mathbf{X} = \mathbf{x} ; \mathbf{\theta}) = \frac{\exp(\lambda(\mathbf{x}))}{1 + \exp(\lambda(\mathbf{x}))}$$ où $$\lambda(\mathbf{\mathbf{x}}) = \theta_0 + \theta_1 x_1 + ... + \theta_K x_K.$$ 

Dans la formule ci-dessus:
- $x_k$ est la cristallisation de la $k^{\text{ème}}$ variable prédictive,
- $\theta_0$ est l'intercept,
- $\theta_k$ est le $k^{\text{ème}}$ coefficient de la régression.

Pour prédire une classe spécifique, il suffit de définir un seuil (threshold) sur les probabilités. Dans scikit-learn, ce seuil est fixé à 0.5 par défaut. Plus précisément, dès que la probabilité prédite est supérieure à 0.5, l'observation est associée à la classe $Y = 1$ (l'étoile géante dans ce cas). Sinon, elle est associée à la classe $Y = 0$ (l'étoile naine).

NB : Bien que cela ne soit pas requis dans cet exercice, il est courant en pratique d’ajuster ce seuil en fonction des spécificités du contexte. Par exemple, dans les applications biomédicales, les conséquences d’une erreur peuvent être coûteuses : détecter un cancer chez une personne en parfaite santé, par exemple. Dans de tels cas, il est pertinent de modifier le seuil pour minimiser le risque d’erreurs critiques.

</div>


````python
# Define the predictive variables and the target variable
X_logistic = data[['B-V', 'Amag']].values
y_logistic = data['TargetClass'].values

# Create and fit the logistic regression model
logistic_model = LogisticRegression()
logistic_model.fit(X_logistic, y_logistic)

# Predict the target variable
y_pred_logistic = logistic_model.predict(X_logistic)

# Calculate the accuracy
accuracy = np.mean(y_pred_logistic == y_logistic)
print(f"Accuracy: {accuracy * 100:.2f}%")
````

   - **Ajout de nouvelles variables prédictives** : Enrichir le modèle avec d'autres caractéristiques stellaires pertinentes pourrait améliorer sa capacité discriminative.

Le modèle actuel fournit des résultats satisfaisants pour une séparation linéaire des classes, mais il présente des limitations dans les zones où les classes se chevauchent ou suivent des relations non linéaires. Des améliorations pourraient être apportées pour mieux gérer ces situations.
